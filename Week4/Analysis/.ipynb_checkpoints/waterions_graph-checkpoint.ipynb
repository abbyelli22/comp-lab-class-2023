```python
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

#ax.fill_between(t,data, color="magenta", linestyle="-", alpha=0.1)
#ax.plot(t,sod_cla_data, color="magenta", linestyle="-", label = "sodium chloride")
ax.plot(t,sod_data, color = 'g', label = "sodium")
ax.plot(t, wateroxygens, color = 'blue', label = "water oxygens")
ax.plot(t, oxygenonly, color = 'red', label = "oxygen only")


ax.set_xlabel("time $t$ (ps)")
ax.set_ylabel(r"MSD (nm\S2\N)")
plt.title("Sodium Chloride, Water Oxygens, and Oxygen MSD vs. time")

plt.legend()

fig.savefig("msd_sod_waterO_O.png", dpi=300)
#fig.savefig("msd_sod_cla.svg")
#fig.savefig("msd_sod_cla.pdf")
```
