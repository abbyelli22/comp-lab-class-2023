```bash
# Gromacs Tutorial
```

```bash
module avail gromacs
```

```bash
module load  gromacs-plumed/openmpi/intel/2022.3 
```

```bash
grep -v HOH 1aki.pdb > 1AKI_clean.pdb
```

```bash
gmx_mpi pdb2gmx -f 1AKI_clean.pdb -o 1AKI_processed.gro -water spce << EOF
15
EOF
```

```bash
gmx_mpi editconf -f 1AKI_processed.gro -o 1AKI_newbox.gro -c -d 1.0 -bt cubic
```

```bash
gmx_mpi solvate -cp 1AKI_newbox.gro -cs spc216.gro -o 1AKI_solv.gro -p topol.top
```

```bash
#Step 4 Adding Ions

# Assemble .tpr file 
gmx_mpi grompp -f ions.mdp -c 1AKI_solv.gro -p topol.top -o ions.tpr
```

```bash
# Pass file to genion
gmx_mpi genion -s ions.tpr -o 1AKI_solv_ions.gro -p topol.top -pname NA -nname CL -neutral << EOF
13
EOF
```

```bash
gmx_mpi grompp -f minim.mdp -c 1AKI_solv_ions.gro -p topol.top -o em.tpr
```

```bash
gmx_mpi mdrun -v -deffnm em
```

```bash
gmx_mpi energy -f em.edr -o potential.xvg << EOF
10 0
EOF
```

```bash
# changed nvt.mdp to run for 50ns instead of 100ns
```

```bash
gmx_mpi grompp -f nvt.mdp -c em.gro -r em.gro -p topol.top -o nvt.tpr
```

```bash
#50 ns
gmx_mpi mdrun -deffnm nvt
```

```bash
gmx_mpi energy -f nvt.edr -o temperature.xvg << EOF
16 0
EOF
```

```bash
gmx_mpi grompp -f npt.mdp -c nvt.gro -r nvt.gro -t nvt.cpt -p topol.top -o npt.tpr
```

```bash
gmx_mpi mdrun -deffnm npt
```

```bash
gmx_mpi energy -f npt.edr -o pressure.xvg << EOF
18 0 
EOF
```
