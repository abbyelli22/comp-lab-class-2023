```python
import matplotlib.pyplot as plt
import numpy 
```

```python
t,data,x,y,z = numpy.loadtxt("gyrate.xvg", unpack=True,comments=['#', '@'])
```

```python
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

ax.fill_between(t,data, color="magenta", linestyle="-", alpha=0.1)
ax.plot(t,data, color="magenta", linestyle="-")

ax.set_xlabel("time $t$ (ps)")
ax.set_ylabel(r"2JOF $R_\mathrm{gyr}$ (nm)")

plt.title("2JOF radius of gyration vs. time")

fig.savefig("rgyr.png", dpi=300)
```

```python
# end to end distance 

t, end_end_dist, x,y,z,a = numpy.loadtxt("polystat.xvg", unpack=True,comments=['#', '@'])

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

ax.fill_between(t,end_end_dist, color="magenta", linestyle="-", alpha=0.1)
ax.plot(t,end_end_dist, color="magenta", linestyle="-")

ax.set_xlabel("time $t$ (ps)")
ax.set_ylabel(r"End to End Distance(nm)")

plt.title("2JOF End to End Distance")

fig.savefig("end_end_dist.png", dpi=300)
```
