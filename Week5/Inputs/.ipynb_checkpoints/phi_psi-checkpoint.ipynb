```python
import mdtraj as md
import numpy 
import matplotlib.pyplot as plt
```

```python
trajectory = md.load("md_adp.trr", top = "adp.gro")
```

```python
phi = md.compute_phi(trajectory)[1].flatten()
print(phi)
```

```python
psi = md.compute_psi(trajectory)[1].flatten()
print(psi)
```

```python
steps = numpy.arange(0,5001,1)
```

```python
time = 0.002 * steps
```

```python
#psi vs time
```

```python
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

ax.plot(psi, label = "psi")
#ax.plot(t,end_end_dist, color="magenta", linestyle="-")

ax.set_xlabel("Time (ps)")
ax.set_ylabel(r"PSI angle")

plt.title("PSI")

fig.savefig("PSI.png", dpi=300)
```

```python
#psi vs time

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

ax.plot(phi, label = "phi")
#ax.plot(t,end_end_dist, color="magenta", linestyle="-")

ax.set_xlabel("Time (ps)")
ax.set_ylabel(r"PHI angle")

plt.title("PHI")

fig.savefig("PHI.png", dpi=300)
```

```python
#Make a free energy surface in phi, psi space from this data. In otherwords, compute a 2d histogram of the phi, psi data compute F(phi,psi)=-log(histogram), and plot F(phi,psi)-min(F(phi,psi)) nicely labeled, with a colorbar, etc
```

```python
print(numpy.array(phi[1]).shape)
```

```python
from matplotlib import pyplot as plt, colors

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
fig,ax = plt.subplots()

min_phi = phi -min(phi)
min_psi = psi - min(psi)

hh = ax.hist2d(min_phi, min_psi, norm = colors.LogNorm())

ax.set_xlabel('Phi')
ax.set_ylabel('Psi')
ax.set_title('Histogram of Phi and Psi Probability')

clb = fig.colorbar(hh[3],ax =ax)
clb.ax.set_title('-log(probability)',fontsize=10)
plt.show
```
