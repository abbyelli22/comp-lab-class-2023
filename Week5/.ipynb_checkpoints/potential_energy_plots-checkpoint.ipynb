```python
import mdtraj as md
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt, colors
```

```python
trajectory = md.load("/home/axe1/comp-lab-class/comp-lab-class-2023/Week5/T300/md_adp.trr", top = "/home/axe1/comp-lab-class/comp-lab-class-2023/Week5/T300/md_adp.gro")
phi = md.compute_phi(trajectory)[1].flatten()
psi = md.compute_psi(trajectory)[1].flatten()

```

```python
#Free energy surface of T300

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
fig,ax = plt.subplots()

min_phi = phi -min(phi)
min_psi = psi - min(psi)

hh = ax.hist2d(min_phi, min_psi, norm = colors.LogNorm())

ax.set_xlabel('Phi')
ax.set_ylabel('Psi')
ax.set_title('Histogram of Phi and Psi Probability')

clb = fig.colorbar(hh[3],ax =ax)
clb.ax.set_title('-log(probability)',fontsize=10)
plt.show
```
