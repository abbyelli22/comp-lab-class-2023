```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time, rep1, rep2, rep3 = np.loadtxt("/home/axe1/comp-lab-class/comp-lab-class-2023/Week5/T300/replica_temp.xvg", unpack = True, comments = ["#", "@"])
```

```python
#replica 1

fig = plt.figure(figsize=(10,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

ax.plot(rep1)

plt.xlim(0,5000)
plt.ylim([-1, 3]) 

ax.set_xlabel("Time (ps)")
ax.set_ylabel("Replica 1")
plt.title("Replica 1 Temp Change")
```
