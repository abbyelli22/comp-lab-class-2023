```python
import os

lammpstrj_list = []
dcd_list = []

equil_dir = "/scratch/work/courses/CHEM-GA-2671-2023fa/students/axe1/comp-lab-class-2023/Week9/Data/equil"
prod_dir = "/scratch/work/courses/CHEM-GA-2671-2023fa/students/axe1/comp-lab-class-2023/Week9/Data/prod"

# Function to filter and sort files in a directory
def filter_and_sort_files(directory, file_extension, prefix):
    files = [filename for filename in os.listdir(directory) if filename.endswith(file_extension)]
    return sorted([prefix + filename for filename in files])

# Add filtered and sorted files to the respective lists
lammpstrj_list = filter_and_sort_files(equil_dir, '.lammpstrj', 'equil/')
dcd_list = filter_and_sort_files(prod_dir, '.dcd', 'prod/')

# Print aligned pairs
for (a, b) in zip(lammpstrj_list, dcd_list):
    print(a, b)
#len(lammpstrj_list)
```

```python
# Radial Distribution function 
temp_list = [0.475, 0.55, 0.5, 0.65, 0.6, 0.7, 0.8, 0.9, 1.5, 1]  

for idx, temp in enumerate(temp_list):
    # Load trajectory and DCD file for the specific temperature
    u = mda.Universe(lammpstrj_list[idx], dcd_list[idx], topology_format="LAMMPSDUMP")

    s1= u.select_atoms('type 1')
    s2 = u.select_atoms('type 2')

    rdf_11 = rdf.InterRDF(s1,s1)
    rdf_11.run()

    rdf_12 = rdf.InterRDF(s1,s2)
    rdf_12.run()

    rdf_22 = rdf.InterRDF(s2,s2)
    rdf_22.run()

    plt.plot(rdf_11.bins, rdf_11.rdf, label= "g_11")
    plt.plot(rdf_12.bins, rdf_12.rdf, label = "g_12")
    plt.plot(rdf_22.bins, rdf_22.rdf, label = "g_22")

    plt.legend()
    plt.title(f"RDF at Temperature {temp}")
    plt.xlabel('Distance')
    plt.ylabel('RDF')
    plt.savefig(f'Figures/rdf_{temp}.png')  # Save figure with temperature-specific name
    plt.close()  # Close the plot to avoid overlap in the next iteration
```

```python
#import MDAnalysis.analysis.msd as msd
temp_list = [0.475, 0.55, 0.5, 0.65, 0.6, 0.7, 0.8, 0.9, 1.5, 1] 
all_msd_data = []
for idx, temp in enumerate(temp_list):
    
    import MDAnalysis.analysis.msd as msd
    # Load trajectory and DCD file for the specific temperature
    u = mda.Universe(lammpstrj_list[idx], dcd_list[idx], topology_format="LAMMPSDUMP")
    
    MSD = msd.EinsteinMSD(u, select='type 1', msd_type='xyz', fft=True)
    MSD.run()
                      
    msd =  MSD.results.timeseries
    nframes = MSD.n_frames
    timestep = 1 # this needs to be the actual time between frames
    lagtimes = np.arange(nframes)*timestep # make the lag-time axis
    
    all_msd_data.append((lagtimes, msd))  # Store lagtimes and msd for each temperature

    
# Plot all MSD data on the same graph
plt.figure(figsize=(8, 6))
for idx, (lagtimes, msd) in enumerate(all_msd_data):
    plt.loglog(lagtimes, msd, label=f"Temperature: {temp_list[idx]}")
    
plt.xlabel("log Lag Time")
plt.ylabel("log MSD")
plt.title("MSD of Type 1 Particles")
plt.legend()
plt.show()
plt.savefig(f'Figures/msd_type1.png')  
    
```

```python
#MSD = 1
temp_list = [0.475, 0.55, 0.5, 0.65, 0.6, 0.7, 0.8, 0.9, 1.5, 1] 
diffusion_times = [] 

for idx, temp in enumerate(temp_list):
    
    import MDAnalysis.analysis.msd as msd
    # Load trajectory and DCD file for the specific temperature
    u = mda.Universe(lammpstrj_list[idx], dcd_list[idx], topology_format="LAMMPSDUMP")
    
    MSD = msd.EinsteinMSD(u, select='type 1', msd_type='xyz', fft=True)
    MSD.run()
                      
    msd =  MSD.results.timeseries
    nframes = MSD.n_frames
    timestep = 1 # this needs to be the actual time between frames
    lagtimes = np.arange(nframes)*timestep # make the lag-time axis
    
    index_msd_1 = np.where(msd >= 1)[0]
    if len(index_msd_1) > 0:
        diff_time = lagtimes[index_msd_1[0]]
        diffusion_times.append(diff_time)  # Store diffusion time for each temperature
    else:
        diffusion_times.append(None)  # Store None if MSD never reaches 1 for this temperature

    
    all_msd_data.append((lagtimes, msd))  # Store lagtimes and msd for each temperature

# Plot diffusion times against temperature on a log-log scale
reciprocal_temps = [1 / temp for temp in temp_list]

# Plot diffusion times against reciprocal temperatures on a log-log scale
plt.figure(figsize=(8, 6))
plt.loglog(reciprocal_temps, diffusion_times, marker='o', linestyle='-')
plt.xlabel("1 / Temperature")
plt.ylabel("Diffusion Time (t where MSD = 1)")
plt.title("Diffusion Time vs 1 / Temperature")
plt.show()

plt.savefig(f'Figures/Diffusion_vs_temp.png')  
```
