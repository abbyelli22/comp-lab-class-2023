```python
import matplotlib.pyplot as plt
import numpy 
import math
import pandas as pd
```

```python
# Energy Minimization
#Energy Minimization step vs. Potential Energy 
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

t,energy = numpy.loadtxt("potential_50ns.xvg", unpack=True,comments=['#', '@'])

ax.plot(t,energy, color = 'g', label = "Potential")

ax.set_xlabel("time $t$ (ps)")
ax.set_ylabel(r"Potential Energy (kJ/mol))")
plt.title("Potential Energy (kJ/mol)")

plt.legend()

fig.savefig("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/potential_energy_plot.png", dpi=300)
```

```python
# Temperature 
# Temp vs. time 
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

t,temp = numpy.loadtxt("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/temperature.xvg", unpack=True,comments=['#', '@'])

ax.plot(t,temp, color = 'g')

ax.set_xlabel("time $t$ (ps)")
ax.set_ylabel(r"Temp (K))")
plt.title("Temp (K)")

plt.legend()
fig.savefig("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/temp_plot.png", dpi=300)
```

```python
# Pressure
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

t,pressure = numpy.loadtxt("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/pressure.xvg", unpack=True,comments=['#', '@'])

pressure = pd.Series(pressure)

#pressure_avg = pressure.rolling(window=100).mean()
#pressure  = numpy.roll(pressure, shift, axis = None)
pressure_avg = pressure.rolling(window=5).mean()

ax.plot(t,pressure, color = 'g')
ax.plot(t, pressure_avg, color = "r", label = "10 ns Runnning Avg")
ax.plot()

ax.set_xlabel("time $t$ (ps)")
ax.set_ylabel(r"Pressure (bar))")
plt.title("Pressure")

plt.legend()
fig.savefig("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/pressure_plot.png", dpi=300)
```

```python
# Density 
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

t,density = numpy.loadtxt("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/density.xvg", unpack=True,comments=['#', '@'])

density = pd.Series(density)

#pressure_avg = pressure.rolling(window=100).mean()
#pressure  = numpy.roll(pressure, shift, axis = None)
density_avg = density.rolling(window=5).mean()

ax.plot(t,density, color = 'g')
ax.plot(t, density_avg, color = "r", label = "10 ns Runnning Avg")
ax.plot()

ax.set_xlabel("time $t$ (ps)")
ax.set_ylabel(r"Density (kg/m^3))")
plt.title("Density")

plt.legend()
fig.savefig("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/density_plot.png", dpi=300)
```

```python
# RMSD
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(111)
fig.subplots_adjust(bottom=0.2)

t,rmsd_equil = numpy.loadtxt("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/50ns/rmsd_50ns.xvg", unpack=True,comments=['#', '@'])
t,rmsd_crys =  numpy.loadtxt("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/50ns/rmsd_50ns_crystal.xvg", unpack=True,comments=['#', '@'])
#rmsd = pd.Series(rmsd)

#pressure_avg = pressure.rolling(window=100).mean()
#pressure  = numpy.roll(pressure, shift, axis = None)
#rmsd_avg = rmsd.rolling(window=5).mean()

ax.plot(t,rmsd_equil, color = 'g', label = "Ref: Equilibrated")
ax.plot(t, rmsd_crys, color = "r", label = "Ref: Crystal")
ax.plot()

ax.set_xlabel("time $t$ (ns)")
ax.set_ylabel(r"RMSD (nm)")
plt.title("RMSD")

plt.legend()
fig.savefig("/home/axe1/comp-lab-class/comp-lab-class-2023/Week3/rmsd_plot.png", dpi=300)
```
