```python
import numpy as np
import matplotlib.pyplot as plt
#load mdtraj
import mdtraj as md
import os
import math
```

#### a) reads in the data file Data/1hz3_T310.run.25000000.energy.xvg

```python
#traj= np.load("Data/1hz3_T310.run.25000000.energy.xvg")
t, Potential, Kinetic_E, Total_E, Temp, Pressure, Volume = np.loadtxt("Data/1hz3_T310.run.25000000.energy.xvg", unpack = True, comments = ['#','@'])
```

```python
list1= [Potential, Kinetic_E, Total_E, Temp, Pressure, Volume]
list1_names = ["Potential", "Kinetic Energy", "Total Energy", "Temp", "Pressure", "Volume"]
list1_titles = dict(zip(list1_names, list1))
```

#### b) Make 6 separate plots for each quantity vs time, with time in ns

```python
for key, value in list1_titles.items():
    fig = plt.figure(figsize=(5,3))
    ax = fig.add_subplot(111)
    fig.subplots_adjust(bottom=0.2)

    ax.plot(t, value)
    ax.set_xlabel("Time (ns)")
    ax.set_ylabel(key)
    ax.set_title("1hz3 "+ key)
    
    #fig.savefig("1hz3_"+key+".png", dpi=300)    
```

### 	c) Make a plot that has kinetic energy and temperature on the same plot, where the left axis is temperature and the right axis is kinetic energy, using different colors and different axis labels


```python
color_temp = "brown"
color_kinetice = "blue"

fig, ax1 = plt.subplots(figsize = (8,8))
ax2 = ax1.twinx()

ax1.plot(t, Temp, color= color_temp)
ax2.plot(t, Kinetic_E, color= color_kinetice)

ax1.set_xlabel("Time (ns)", fontsize= 15)
ax1.set_ylabel("Temp", color = color_temp, fontsize=15)
ax1.tick_params(axis = "y", labelcolor= color_temp)

ax2.set_ylabel("Kinetic Energy", color = color_kinetice, fontsize= 15)
ax2.tick_params(axis="y", labelcolor= color_kinetice)

fig.suptitle("1hz3 Temp and Kinetic Energy", fontsize = 20)
fig.autofmt_xdate()
```
